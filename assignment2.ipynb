```python
import seaborn as sns
import pandas as pd


# Load training and test data
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')



print(train_data['OverallQual'].unique())
print(train_data['OverallCond'].unique())
print(train_data['YearBuilt'].unique())

print(test_data['OverallQual'].unique())
print(test_data['OverallCond'].unique())
print(test_data['YearBuilt'].unique())

titanic=pd.read_csv('train.csv')
titanic.head(10)

```

```python
# Method 2
# Replace 'Missing' in 'YearBuilt' with its median
year_built_median = train_data['YearBuilt'][train_data['YearBuilt'] != 'Missing'].astype(int).median()
train_data['YearBuilt'].replace('Missing', str(year_built_median), inplace=True)
test_data['YearBuilt'].replace('Missing', str(year_built_median), inplace=True)



# Replace 'Missing' in 'OverallQual' with its mode
overall_qual_mode = train_data['OverallQual'][train_data['OverallQual'] != 'Missing'].mode()[0]
train_data['OverallQual'].replace('Missing',  str(overall_qual_mode), inplace=True)
test_data['OverallQual'].replace('Missing',  str(overall_qual_mode), inplace=True)


# Replace 'Missing' in 'OverallCond' with its mode
overall_cond_mode = train_data['OverallCond'][train_data['OverallCond'] != 'Missing'].mode()[0]
train_data['OverallCond'].replace('Missing',  str(overall_cond_mode), inplace=True)
test_data['OverallCond'].replace('Missing',  str(overall_cond_mode), inplace=True)


print(f"Replaced 'Missing' values in 'YearBuilt' with: {year_built_median}")
print(f"Replaced 'Missing' values in 'OverallQual' with: {overall_qual_mode}")
print(f"Replaced 'Missing' values in 'OverallCond' with: {overall_cond_mode}")


print(train_data['OverallQual'].unique())
print(train_data['OverallCond'].unique())
print(train_data['YearBuilt'].unique())

print(test_data['OverallQual'].unique())
print(test_data['OverallCond'].unique())
print(test_data['YearBuilt'].unique())




```

```python
# Method 1
# Replacing any 'Missing' values in the train and test  with the number 1
train_data.replace('Missing', 1, inplace=True)
test_data.replace('Missing', 1, inplace=True)


print(train_data['OverallQual'].unique())
print(train_data['OverallCond'].unique())
print(train_data['YearBuilt'].unique())

print(test_data['OverallQual'].unique())
print(test_data['OverallCond'].unique())
print(test_data['YearBuilt'].unique())
```

```python
# List of categorical features
categorical_features_all = ['MSSubClass', 'Street', 'HouseStyle', 'ExterQual', 'BldgType', 'LotConfig']
train_data = pd.get_dummies(train_data, columns=categorical_features_all, drop_first=True)
test_data = pd.get_dummies(test_data, columns=categorical_features_all, drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Converting data types
train_data['OverallQual'] = train_data['OverallQual'].astype(str)
test_data['OverallQual'] = test_data['OverallQual'].astype(str)

#  using LabelEncoder
le = LabelEncoder()
train_data['OverallQual'] = le.fit_transform(train_data['OverallQual'])
test_data['OverallQual'] = le.transform(test_data['OverallQual'])
```

```python
from sklearn.preprocessing import StandardScaler
# List of numerical features to be scaled
numerical_features_all = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold']
# using StandardScaler
scaler = StandardScaler()
train_data[numerical_features_all] = scaler.fit_transform(train_data[numerical_features_all])
test_data[numerical_features_all] = scaler.transform(test_data[numerical_features_all])

```

```python
# Align train and test data to have the same columns, filling missing values in test data with 0
train_data, test_data = train_data.align(test_data, join='left', axis=1)
test_data.fillna(0, inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np


X_train = train_data.drop('SalePrice', axis=1)  # Features for the training
y_train = train_data['SalePrice']  # Target variable for the training
X_test = test_data.drop('SalePrice', axis=1)# Features for the test
y_test = test_data['SalePrice'] # Target variable for the test

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_predictions = lr.predict(X_test)
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_predictions))


# Random Forest Regression
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_predictions = rf.predict(X_test)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_predictions))


print(f"Linear Regression RMSE: {lr_rmse}")
print(f"Random Forest Regression RMSE: {rf_rmse}")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Linear Regression scatter plot
sns.scatterplot(x=y_test, y=lr_predictions, ax=ax[0], color='blue', alpha=0.6)
ax[0].set_title('Linear Regression Predictions')
ax[0].set_xlabel('Actual Values')
ax[0].set_ylabel('Predicted Values')
ax[0].plot(y_test, y_test, color='red', label='Ideal Regression Line')
ax[0].legend()

# Random Forest Regression scatter plot
sns.scatterplot(x=y_test, y=rf_predictions, ax=ax[1], color='green', alpha=0.6)
ax[1].set_title('Random Forest Regression Predictions')
ax[1].set_xlabel('Actual Values')
ax[1].set_ylabel('Predicted Values')
ax[1].plot(y_test, y_test, color='red', label='Ideal Regression Line')
ax[1].legend()


plt.tight_layout()
plt.show()

```
